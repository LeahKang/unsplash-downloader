```python
def clean_tags(tags):
    tags = ast.literal_eval(tags)
    #print(tags)
    results = [ i['title'] for i in tags ]
    return results

metadata['tags(clean)'] = metadata['tags'].apply(clean_tags)
metadata['tags(clean)']
```

```python
tags_list = []
for tags in metadata['tags(clean)']:
    tags_list.extend(tags)
    
#tags_list
```

```python
pos_all = dict()
for w in tags_list:
    pos_l = set()
    for tmp in wn.synsets(w):
        if tmp.name().split('.')[0] == w:
            pos_l.add(tmp.pos())
    pos_all[w] = pos_l
unsplash_pos = pos_all
```
